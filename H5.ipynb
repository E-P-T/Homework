# Task 4.2
Implement a function which search for most common words in the file.
Use `data/lorem_ipsum.txt` file as a example.

```python
def most_common_words('lorem_ipsum.txt', number_of_words=3):

    """ a function which search for most common words in the file """

    with open('lorem_ipsum.txt', 'r') as fd:
        words_lst = fd.read().replace('.', '').replace(',', '').lower().split()
    words_dict = dict()
    for word in words_lst:
        words_dict[word] = words_dict.get(word, 0) + 1
    sorted_words_lst = sorted(words_dict, reverse=True, key=lambda key: words_dict[key])
    common_words = list()
    for key in sorted_words_lst:
        if len(common_words) == number_of_words:
            return common_words
        common_words.append(key)
    return common_words
```

```python
import csv

def get_third_elem(iterable):
    return iterable[2]
def get_second_elem(iterable):
    return iterable[1]
def get_first_item(iterable):
    return iterable[0]
def get_top_performers(file_path = 'students.csv', number_of_top_students=5):
    data = csv.reader(open(file_path), delimiter=',')
    g = sorted(data, key= get_third_elem)
    a = g[-number_of_top_students-1:-1]
    a.reverse()
    s = list(map(get_first_item, a))
    return s
print(get_top_performers())
```

```python
def writeSortedStudents(file_path = 'students.csv'):
    data = csv.reader(open(file_path), delimiter=',')
    g = sorted(data, key=get_second_elem)
    g.reverse()
    # print(g)
    with open('sortedStudents.csv', 'w', encoding='UTF8') as f:
        writer = csv.writer(f)
        writer.writerows(g)
writeSortedStudents()
```

### Task 4.5
Implement a decorator `remember_result` which remembers last result of function it decorates and prints it before next call.

```python
def LastCall(func):
	a = None
	def Func(*args):
		nonlocal a
		print(a)
		a = func(*args)
	return Func

@LastCall
def sum_of_list(*args):
	result = ""
	for item in args:
		result += item
	print(f"Current result = '{result}'")
	return result
sum_of_list("asdfghjkl","gfdsa","hgfd")
sum_of_list("aa","bb","cc")
sum_of_list("1","kjj","jh")
```

### Task 4.6
Implement a decorator `call_once` which runs a function or method once and caches the result.
All consecutive calls to this function should return cached result no matter the arguments.

```python
def call(func):
    a = None
    def inner(*args):
        nonlocal a
        if a == None:
            a=func(*args)
            return a
        else:
            return a
    return inner
@call
def sum_of_numbers(a, b):
    return a + b
print(sum_of_numbers(2,3))
print(sum_of_numbers(3,4))
print(sum_of_numbers(5,6))
print(sum_of_numbers(7,8))
```
